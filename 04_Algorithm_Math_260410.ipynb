```python
def prime(n):
    primes = [2]
    for i in range(3,n+1,2):
        isprime = 1
        for j in primes:
            if i%j==0:
                isprime = 0
        if isprime==1:
            if i not in primes:
                primes.append(i)
    return primes
def is_prime_number(n):
    primes = prime(n)
    if n in primes:
        return True
    else:
        return False
print(is_prime_number(1))
```
