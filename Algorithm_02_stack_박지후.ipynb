```python
#데이터 정의
stack = []
while True:
    #메뉴를 출력
    print('''
    ===================================================================================
    1.push 2.pop 3.peek 0.exit
    ===================================================================================
    ''')
    #입력을 받는다
    menu=int(input('>>연산 메뉴 선택:'))
    #print(menu)
    #입력에 따라 stack연산을 수행하고
    if menu == 0:
        break
    elif menu == 1: 
        data = int(input('[Push] 추가할 데이터 입력 : '))
        stack.append(data)
        print('[Stack]',stack)
    elif menu ==2: 
        data = stack.pop()
        print('[Pop] 가져온 데이터 :',data)
    elif menu == 3:
        data= stack[len(stack)-1]
        print('top 데이터 확인 :',data)
    #스택의 상태를 출력한다
    #위의 과정을 0.exit을 선택할 떄까지 반복
    print('[Stack]',stack)
print('정수형 스택 종료')
```
