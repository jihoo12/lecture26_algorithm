```python
#파이썬 리스트 연습: 항목 삽입과 삭제
data_list = [23,45,21,14,2,67]
data_list.append(1)
print(data_list)
data_list.pop(1)
print(data_list)
```

```python
#리스트와 반복구조
#주어진 데이터가 어디에 저장되어있는지 출력
data_list = [23,45,21,14,2,67]
target = 14
index = 0
while index <= 5:
    if data_list[index] == target:
        break
    index += 1
print(index)
```

```python
for index in range(6):
    if data_list[index] == target:
        break
print(index)
```

```python
#리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)
data_list = [23,45,21,14,2,67]
max = data_list[0]
max_index = 0
for index in range(1,6):
    if data_list[index] > max:
        max = data_list[index]
        max_index = index
print(max,max_index)
```

```python
data_list = [23,45,21,14,2,67]
max = data_list[0]
for data in data_list:
    if data > max:
        max = data
print(max)
```

```python
#주어진 데이터가 어디에 저장되어 있는지 출력
data_list = [23,45,21,14,2,67]
target = 67
count =0
for i in data_list:
    if i == target:
        print(count)
    else:
        count += 1
```

```python
#리스트에 저장된 숫자 중 가장 작은 숫자 출력 (최소값)
data_list = [23,45,21,14,2,67]
min = data_list[0]
min_index = 0
for index in range(1,6):
    if data_list[index] < min:
        min = data_list[index]
        min_index = index
print(min,min_index)
```

```python
#리스트에 저장된 숫자 중 입력된 숫자보다 큰 수가 몇 개 있는지 구하기
data_list = [23,45,21,14,2,67]
a = int(input())
c = 0
for index in range(1,6):
    if data_list[index] > a:
        c += 1
print(c)
```

```python
#리스트에 저장된 모든 값 더하여 출력(합)
data_list = [23,45,21,14,2,67]
temp = 0
for i in data_list:
    temp = temp + i
print(temp)
```

```python
#리스트에 저장된 숫자를 역순으로 저장하여 출력
data_list = [23,45,21,14,2,67]
print(data_list[::-1])
```
