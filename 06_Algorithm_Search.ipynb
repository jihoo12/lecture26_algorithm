```python
# recursive call @ binary search
def binary_search_recursive(A, key, low, high):
    if (low > high):
        return -1 
    middle = (low + high) // 2
    if key == A[middle]: return middle
    elif key < A[middle]: 
        return binary_search_recursive(A, key, low, middle-1)
    else:
        return binary_search_recursive(A, key, middle+1, high)


```

```python
def binary_search(A, key):
    low = 0
    high = len(A) - 1
    
    while low <= high:
        middle = (low + high) // 2  # 중간 인덱스 계산
        
        if A[middle] == key:        # 탐색 성공
            return middle
        elif A[middle] > key:      # 키 값이 중간보다 작으면 왼쪽 탐색
            high = middle - 1
        else:                      # 키 값이 중간보다 크면 오른쪽 탐색
            low = middle + 1
            
    return -1                      # 탐색 실패
```

```python
data = [2,6,11,13,18,20,22,27,29,30,34,38,41,42,45,47]
num = int(input('찾을 숫자:'))
index = binary_search_recursive(data,num,0,len(data))
if index== -1:
    print('없는 숫자입니다')
else:
    print(f'>>> 숫자{num}은 index {index}에 있습니다')
index = binary_search(data,num)
if index == 1:
    print('없는 숫자입니다')
else:
    print(f'>>> 숫자{num}은 index{index}에 있습니다')
```
